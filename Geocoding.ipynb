```python
import pandas as pd

df_hougang = pd.read_csv("HDB_Hougang_Final.csv")
df_serangoon = pd.read_csv("HDB_Serangoon_Final.csv")

def nearest_mrt(location, mrt_list):
    from geopy import distance
    min_dist = 1000
    mrt = ""
    for i in mrt_list:
        mrt_coor = (i[1], i[2])
        cal_dist = distance.distance(location, mrt_coor).km
        if cal_dist < min_dist:
            min_dist = cal_dist
            mrt = i[0]
    return (mrt, min_dist)

mrt_hougang = []
distance_hougang = []
mrt_serangoon = []
distance_serangoon = []

for i in range(len(df_hougang)):
    res_loc = (df_hougang["X_coordinates"][i], df_hougang["Y_coordinates"][i])
    result = nearest_mrt(res_loc, mrt_loc)
    mrt_hougang.append(result[0])
    distance_hougang.append(result[1])
    if i%100 == 0:
        print(i)
        
for i in range(len(df_serangoon)):
    res_loc = (df_serangoon["X_coordinates"][i], df_serangoon["Y_coordinates"][i])
    result = nearest_mrt(res_loc, mrt_loc)
    mrt_serangoon.append(result[0])
    distance_serangoon.append(result[1])
    if i%100 == 0:
        print(i)
```
